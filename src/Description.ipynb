```python
from bs4 import BeautifulSoup
import requests
import pandas
from openpyxl import load_workbook
```

```python
wb = load_workbook(filename = 'links.xlsx')
print(wb)
```

```python
sheet_ranges = wb['sheet1']
print(sheet_ranges)
```

```python
places = []
for row in sheet_ranges.iter_rows(): 
    for cell in row:
        places.append(cell.value)
print(places)
print(places[0].replace(' ','_'))
```

```python
webLinks = []
for i in range(len(places)):
    if places[i] != 'None':
        places[i] = places[i].replace(" ","_")
        #print(link)
        url = "https://en.wikipedia.org/wiki/"+places[i]
        webLinks.append(url)
```

```python
#webLinks[] = ''

for index,url in enumerate(webLinks):
    print(url)
```

```python
location_list = []

def getLocation(url):
    if url !='nn':
        web_text = requests.get(url).text
        web_soup = BeautifulSoup(web_text,'lxml')
        table = web_soup.find('table', class_='infobox')
        if table:
            tr_list = table.find_all('tr')
            for tr in tr_list:
                if tr.th and (tr.th.text == 'Location' or tr.th.text == 'Province' or tr.th.text == 'District' or tr.th.text == 'Town or city'):
                    if tr.td:
                        location = tr.td.text
                        location_list.append(location.strip())
                        return location
    else:
        location_list.append('none')        
```

```python
for index,url in enumerate(webLinks):
    print(getLocation(url))
```

```python
def getDescription(url):
    desc=''
    flag = 1
    if url =='none' or url == 'nn':
        flag = 0
    else:
        html_text = requests.get(url).text
        soup = BeautifulSoup(html_text, 'lxml')
        p_tags = soup.find_all('p')
        
        i = soup.find('p',class_='mw-empty-elt')
        if i in p_tags:
            p_tags.remove(i)
        
        for txt in p_tags[0].find_all(text=True):
            desc += txt
        return(desc)

    if flag==0:
        return('none')
```

```python
for index,url in enumerate(webLinks):
    print(getDescription(url))
```
